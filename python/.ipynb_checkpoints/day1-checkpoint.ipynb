```python
#Even or Odd Checker

num = int(input("Enter number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#Sum of First 10 Numbers

total = 0

for i in range(1, 11):
    total += i

print(total)
```

```python
#mini task
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```
